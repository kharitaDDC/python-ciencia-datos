```python
df_sales = df_sales.dropna(subset=['distributor'])
# cleaned_data.to_csv("cleaned_sales_in_paraguay.csv", index=False)
```

```python
df_sales.info()
```

```python
# CREO Lista de columnas df_sales

list_col = list(df_sales.columns[1:13])
list_col
```

```python
for col in list_col:
  df_sales[col] = df_sales[col].str.replace("$","")
  df_sales[col] = df_sales[col].str.replace(".","")   # Comas por puntos, no al revez , python trabaja asi los numeros, con punto decimal
  df_sales[col] = df_sales[col].str.replace(",",".")
df_sales.head()
```

```python
# PASAR a "float"
for col in list_col:
  df_sales[col] = df_sales[col].astype(float)


#df['id pais'] = df['id pais'].astype(int)
df_sales.dtypes
```
