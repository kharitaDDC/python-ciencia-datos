### Detectar Valores Ausente  y Valores Atipicos .Decidir si eliminarlos y el por que de la eleccion.


```python
# Calificaciones
print(data.head())
print(data.shape)
print(data.info())
```

```python
# Obtener el número de valores faltantes en cada columna
data.isnull().sum()
```

```python
# Descripcion del dataframe
data.describe()
```

```python
# Verificamos si hay valores atipicos u Outlayers
data[['gender', 'race/ethnicity', 'parental level of education', 'lunch',
       'test preparation course', 'math score', 'reading score',
       'writing score']].boxplot()
plt.show()
```

```python
# Vemos una concentracion pareja en torno a la mediana en cada Materia y se observan valores atipicos en cada uno
# Se evalua la limpieza de los mismos
```

```python
# Elimina valores atípicos utilizando el método IQR

#'math score', 'reading score','writing score']
Q1m = data['math score'].quantile(0.25)
Q3m = data['math score'].quantile(0.75)
IQRm = Q3m - Q1m
data = data[(data['math score'] >= Q1m - 1.5 * IQRm) & (data['math score'] <= Q3m + 1.5 * IQRm)]

Q1r = data['reading score'].quantile(0.25)
Q3r = data['reading score'].quantile(0.75)
IQRr = Q3r - Q1r
data = data[(data['reading score'] >= Q1r - 1.5 * IQRr) & (data['reading score'] <= Q3r + 1.5 * IQRr)]

Q1w = data['writing score'].quantile(0.25)
Q3w = data['writing score'].quantile(0.75)
IQRw = Q3w - Q1w
data = data[(data['writing score'] >= Q1w - 1.5 * IQRw) & (data['writing score'] <= Q3w + 1.5 * IQRw)]


```

```python
# Verificamos si hay valores atipicos u Outlayers ya no estan
data[['gender', 'race/ethnicity', 'parental level of education', 'lunch',
       'test preparation course', 'math score', 'reading score',
       'writing score']].boxplot()
plt.show()
```

# **Ejercicio 6**\
## Realizar analisis de matriz corelacion y explicar que variable estan correlacionadas\

```python
# Matriz de correlación de Pearson
data_mat = data[['math score', 'reading score',
       'writing score']]
correlacion = np.round(data_mat.corr(), 2)
print(correlacion)
print(np.round(np.cov(data_mat, rowvar=False), 2))  # Covarianza
```

```python
# Graficamos los datos de correlaciones en un mapa de calor
sns.heatmap(correlacion, annot=True, cmap='coolwarm')
plt.show()
```

```python
### .Vemos una correlacion alta de 0.95,   entre  wirting_score y reading_score
```
